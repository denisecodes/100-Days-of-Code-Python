```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

```python
# Insert spread_col at position 1, name: Spread
```

```python
clean_df.insert(1, 'Spread', spread_col)
```

```python
clean_df.head()
```

```python
# Sort values by the Spread column from lowest to highest
```

```python
low_risk = clean_df.sort_values('Spread')
```

```python
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# Top 5 degrees with the highest values in the 90th percentile
```

```python
top_salaries = clean_df.sort_values(by=['Mid-Career 90th Percentile Salary'], ascending=False)
top_salaries[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# Majors with the largest difference between high and low earners after graduation
```

```python
high_risk = clean_df.sort_values(by=['Spread'], ascending=False)
```

```python
high_risk[['Undergraduate Major', 'Spread']].head()
```
